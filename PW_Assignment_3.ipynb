# 6. Explain the concept of generators in Python and how they are defined.

Generators are a type of iterable, like lists or tuples, but instead of storing all values in memory, they generate values on the fly using the yield keyword.

A generator function is defined like a regular function but uses yield to return values one at a time, pausing between each.

```python
#Example :

def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1
for num in count_up_to(3):
    print(num)
```

# 7. What are the advantages of using generators over regular functions?

* Memory Efficient: They generate values one by one, so they don’t use memory to store all items.

* Faster for large data: No need to wait for the whole list to be created.

* Lazy Evaluation: Values are computed only when needed.


```python
#Example:

def big_generator():
    for i in range(1000000):
        yield i

result = print(big_generator())
```

Syntax:

# 10. What is the difference between map(), reduce(), and filter() functions in Python?

* map(): Applies a function to every element in the iterable.

* filter(): Filters elements based on a condition.

* reduce(): Applies a function to pairs of elements cumulatively to reduce the iterable to one value (requires functools).


```python
#Example:

from functools import reduce

nums = [1, 2, 3, 4]

#map
squared = list(map(lambda x: x**2, nums))

# filter
even = list(filter(lambda x: x % 2 == 0, nums))

# reduce
total = reduce(lambda x, y: x + y, nums)

print(squared)
print(even)
print(total)
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

for i in range(1, 201):
    if is_prime(i):
        print(i, "is prime")
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class Fibonacci:
    def __init__(self, n):
        self.n = n
        self.a = 0
        self.b = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

fib = Fibonacci(10)
for num in fib:
    print(num)
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(max_exponent):
    for i in range(max_exponent + 1):
        yield 2 ** i

for power in powers_of_two(5):
    print(power)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(s):
    return ''.join(filter(lambda x: x.lower() not in 'aeiou', s))

print(remove_vowels("Hello World"))
```
